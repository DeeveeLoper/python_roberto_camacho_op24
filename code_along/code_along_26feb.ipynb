Pandas aggregation, group by

if i query a taxis table in a database with sql, i can do this:

SELECT AVG(fare) as avf_fare, AVG(distance) as avg_distance
FROM taxis
GROUP BY payment


```python
df.groupby("payment")[["fare", "distance"]].mean()
```

```python
df.groupby("payment").agg(dict(fare = "mean", distance= "max"))
```
