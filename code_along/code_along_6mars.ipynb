```python
import pandas as pd

url = "https://en.wikipedia.org/wiki/2024_Summer_Olympics_medal_table"

#df_country = pd.read_html(url)[3]
df_country = pd.read_html(url, index_col=0)[3]

df_country.index.name = None

#type(df_country)

#df_country.head()

df_country = df_country[:-1]

df_country["NOC"] = df_country["NOC"].str.replace(r'[^A-Za-z\s]', "", regex=True)
#df_country["NOC"] = df_country["NOC"].str.replacer("*", "", regex=True)

#df_country["NOC"].unique()

df_country
```
