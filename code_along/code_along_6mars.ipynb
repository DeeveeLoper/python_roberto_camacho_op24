```python
import pandas as pd

url = "https://en.wikipedia.org/wiki/2024_Summer_Olympics_medal_table"

#df_country = pd.read_html(url)[3]
df_country = pd.read_html(url, index_col=0)[3]

df_country.index.name = None

#type(df_country)

#df_country.head()

df_country = df_country[:-1]

df_country["NOC"] = df_country["NOC"].str.replace(r'[^A-Za-z\s]', "", regex=True)
#df_country["NOC"] = df_country["NOC"].str.replacer("*", "", regex=True)

#df_country["NOC"].unique()

df_country
```

```python
df_country.info()
```

```python
fig_country = df_country.plot(kind="bar", x = "NOC", y = "Total")
```

```python
df_sorted = df_country.head(8).sort_values("Total", ascending=False)

nr_top_countries = 8
red, gray = "#ff0000", "#4b5563"
colors = [red]*3 + [gray]*(nr_top_countries-3)

fig_country = df_sorted.plot(kind="bar", 
                             x = "NOC", 
                             y = "Total", 
                             color = colors, rot= 70
                             )

fig_country.legend().set_visible(False)

fig_country.set_xlabel("Top 8 countries", loc="left", color=gray)

fig_country.set_title("Total number of medals", loc="center", color=gray, fontsize=20)
```
